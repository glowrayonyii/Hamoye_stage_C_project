             DATA CLEANING

```python
pd.set_option('display.max_rows', 50)
df
```

```python
#using regex to replace the parenthesis around the Release_year
import re
df['Release_Year'] = df['Release_Year'].str.replace(r'[()]', "")
df
```

```python
#more cleaning process on the release year column
df['Release_Year'] = df['Release_Year'].str.replace('I', '').str.replace('II', '')
df['Runtime'] = df['Runtime'].str.replace('min', '')
df
```

```python
#stripping off whitespaces
df['Movie_name'].str.strip()
df['Genre'].str.strip()
df['Release_Year'].str.strip()
df['Runtime'].str.strip()
df
```

```python
#changing the data type
df['Movie_name'] = df['Movie_name'].astype('string')
df['Genre'] = df['Genre'].astype('string')
df['Release_Year'] = df['Release_Year'].astype('int')
df['Runtime'] = df['Runtime'].astype('int')
df.info()
```

            ANALYSIS

```python
#What is the top ten rated movies?
top_ten = df.groupby('Movie_name')['IMDB_Ratings'].max().sort_values(ascending = False)
top_ten.head(10)
```

```python
#Creating a table plot
top_ten = df.sort_values('IMDB_Ratings', ascending=False).head(10)

fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')
table = ax.table(cellText = top_ten.values, colLabels = top_ten.columns, loc ='center')
table.auto_set_font_size(False)
table.set_fontsize(8)
table.scale(1.5, 1.2)

plt.title('Top Ten Highest Movie Ratings')
plt.show()
```

```python
#average user ratings over time rounded in 2 decimal places
over_time = df.groupby('Release_Year')['IMDB_Ratings'].mean().round(2).reset_index()
plt.figure (figsize=[10, 4])
plt.plot(over_time['Release_Year'], over_time['IMDB_Ratings'], marker = 'o')
plt.xlabel('Release_Year')
plt.ylabel('IMDB_Ratings')
plt.title('Average User Ratings per Year')
plt.style.use('ggplot')
plt.savefig('overtime.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
#import libraries for scraping
import requests
from bs4 import BeautifulSoup
```

```python
#fetch webpage
url = 'https://www.imdb.com/search/title/?groups=top_1000&sort=release_date,asc&start=701&ref_=adv_nxt'
response = requests.get(url)
content = response.content
content
```

```python
#create BeautifulSoup object
soup = BeautifulSoup(content, "html.parser")
```

```python
#find the container to extract the needed information
titles = soup.findAll('div', 'lister-item-content')
movies = [title.find('a').text
        for title in titles]
movies
```

```python
release = [title.find('span', 'lister-item-year text-muted unbold').text.strip()
          for title in titles]
release
```

```python
genre = [title.find('span', 'genre').text.strip()
        for title in titles]
genre
```

```python
runtime = [title.find('span', 'runtime').text.strip()
          for title in titles]
runtime
```

```python
ratings = [title.find('div', 'inline-block ratings-imdb-rating', text_attribute = False).text.strip()
          for title in titles]
ratings
```
